```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("DailyDelhiClimateTrain.csv")
```

```python
data['date'] = pd.to_datetime(data.date)
```

```python
fig = px.line(data,x='date',y='meantemp',title="Mean Temperature In Delhi Over the years")
fig.show()
```

```python
fig = px.line(data,x='date',y='humidity',title="Humidity in Delhi over the years")
fig.show()
```

```python
fig = px.line(data,x='date',y='wind_speed',title="Wind Speed in Delhi over the years")
fig.show()
```

```python
px.scatter(data,x="humidity",y="meantemp",trendline='ols',title = "Relationship between temperature and humidity")
```

```python
data.columns
```

```python
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

```python
forecast_data = data.rename(columns= {
    'date' : 'ds',
    'meantemp' : 'y'
})
print(forecast_data)
```

```python
from prophet import Prophet
from prophet.plot import  plot_plotly , plot_components_plotly

model = Prophet()

model.fit(forecast_data)
```

```python
forecasts = model.make_future_dataframe(periods=365)
```

```python
predictions = model.predict(forecasts)
```

```python
plot_plotly(model , predictions)
```

```python
plot_components_plotly(model , predictions)
```
