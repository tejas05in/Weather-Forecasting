```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("DailyDelhiClimateTrain.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['date'] = pd.to_datetime(data.date)
```

```python
fig = px.line(data,x='date',y='meantemp',title="Mean Temperature In Delhi Over the years")
fig.show()
```

```python
fig = px.line(data,x='date',y='humidity',title="Humidity in Delhi over the years")
fig.show()
```

```python
fig = px.line(data,x='date',y='wind_speed',title="Wind Speed in Delhi over the years")
fig.show()
```

```python
px.scatter(data,x="humidity",y="meantemp",title = "Relationship between temperature and humidity")
```
